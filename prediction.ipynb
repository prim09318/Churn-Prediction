```python
#we will learn to load pkl files and models 
# how we can convert new data in raw format to our desired format

import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
## load trained model,scaler ppickle, and onehot encoded files
model=load_model('model.h5')
```

```python
## loading the encoder and scaler
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open ('onehot_encoder_geo.pkl','rb') as file:
    onehot_encoder_geo=pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
##Example input
input_data={
    'CreditScore':608.00,
    'Geography':'Germany',
    'Gender':'Male',
    'Age':41.00,
    'Tenure':1.00,
    'Balance':83807.86,
    'NumOfProducts':1.00,
    'HasCrCard':0.00,
    'IsActiveMember':1.00,
    'EstimatedSalary':112542.58,
}
```

```python
#one hot encode Geography
geo_encoded=label_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=label_encoder_geo.get_feature_names_out(['Geography']))
```

```python
## converting input_data from key value pairs to dataframe
input_df=pd.DataFrame([input_data])
```

```python
input_df
```

```python
## encode categorical variables
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
```

```python
## concatination onehot encoded
input_df=pd.concat([input_df.drop("Geography",axis=1),geo_encoded_df],axis=1)
```

```python
input_df
```

```python
## scaling the input data
input_scaled=scaler.transform(input_df)
```

```python
##Predicting churn
prediction=model.predict(input_scaled)
prediction_proba=prediction[0][0]
prediction_proba
```

```python
prediction
```

```python
if prediction_proba>0.5:
    print('The customer is likely to churn')
else:
    print('The customer is not likely to churn')
```
